```python
numerical_cols = [col for col in data.columns if data[col].dtype == 'int64']
len(numerical_cols)
```

```python
categorical_cols = [col for col in data.columns if data[col].dtype == 'object']
len(categorical_cols)
```

```python
import matplotlib.pyplot as plt

# Sample DataFrame
# Assuming 'data' is your DataFrame
# Create subplots
fig, axes = plt.subplots(len(numerical_cols), 1, figsize=(15, 5 * len(numerical_cols)))

# Plot boxplots for each column
for i, col in enumerate(numerical_cols):
        data.boxplot(column=col, ax=axes[i])
        axes[i].set_title(col)
        
    

plt.tight_layout()
plt.show()
```

```python
for col in categorical_cols:
    result = data[col].value_counts()
    print(result)
    print('-----------------')
```

```python
import matplotlib.pyplot as plt

# Sample DataFrame
# Assuming 'data' is your DataFrame

# Calculate the number of rows needed for subplots
num_rows = len(categorical_cols)

# Create subplots
fig, axes = plt.subplots(num_rows, 1, figsize=(15, 5 * num_rows))

# Plot histograms for each column
for i, col in enumerate(categorical_cols):
    data[col].hist(bins=data[col].nunique(), ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

```python
bins = pd.IntervalIndex.from_tuples([(18, 25), (25, 32), (32, 39), (39, 46), (46, 53),(53,60)])
Age = data['Age']
data['Age'] = pd.cut(Age, bins)
data['Age']
```

```python
bons = pd.IntervalIndex.from_tuples([(0, 5), (5, 10), (10, 15), (15, 20), (20, 25),(25,30),(30,35),(35,40)])
TotalWorkingYears = data['TotalWorkingYears'] 
data['TotalWorkingYears'] = pd.cut(TotalWorkingYears, bons)
data['TotalWorkingYears']
```
