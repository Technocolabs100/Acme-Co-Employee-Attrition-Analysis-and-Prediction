```python
# Create a pivot table for the plot
pivot_data = data.pivot_table(index='Department', columns='Attrition', aggfunc='size', fill_value=0)

# Plot stacked bar plot
pivot_data.plot(kind='bar', figsize=(12, 5))
plt.title("Attrition based on Department")
plt.ylabel("Number of Employees")
plt.show()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoding_cols = ['BusinessTravel','JobRole','EducationField','Department','MaritalStatus']
label_encode = {}
for col in encoding_cols:
     label_encode[col] = LabelEncoder() # The process is identify the unique integer for each categories.
     data[col] = label_encode[col].fit_transform(data[col])
```
