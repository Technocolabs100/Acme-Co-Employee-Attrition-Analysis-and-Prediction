___1.8.2  Correlation Between Numerical Features and Target :___
 - _Using Point-Biserial Correlation Method_

```python
#Point-Biserial Correlation
pb_corr = {}
for feature in filtered_numerical_columns: 
    corr, _ = pointbiserialr(df[feature],df['Attrition'].replace({"Yes":1,"No":0}))
    pb_corr[feature] = corr
#--------------------------------------------------------------------------------------------------------------
numerical_features_target_correlation=pd.DataFrame(pd.Series(pb_corr)).transpose()
#--------------------------------------------------------------------------------------
#Bar graph
plt.figure(figsize=(22, 5))
sns.barplot(data=numerical_features_target_correlation, color='#144bff')
plt.xlabel('Numerical Features')
plt.ylabel('Point-Biserial Correlation')
plt.title('Point-Biserial Correlation for Numerical Features vs Attrition')
plt.xticks(rotation=90)
plt.show()
#--------------------------------------------------------------------------------------------------------------
numerical_features_target_correlation
```

___1.8.3  Correlation Between Categorical Features and Target :___
 - _Using Chi-Square Test_

```python
#Chi-square test
chi2_test = {}
for cat_feature in categorical_columns[1:]:  
    contingency_table = pd.crosstab(df[cat_feature], df['Attrition'])
    chi2, _, _, _ = chi2_contingency(contingency_table)
    chi2_test[cat_feature] = chi2
#--------------------------------------------------------------------------------------------------------------
category_feature_target_correlation=pd.DataFrame(pd.Series(chi2_test)).transpose()
#--------------------------------------------------------------------------------------------------------------
#Bar graph
plt.figure(figsize=(22, 4))
sns.barplot(data=category_feature_target_correlation, color='#144bff')
plt.xlabel('Categorical Features')
plt.ylabel('Chi-square Test Statistics')
plt.title('Chi-square Test Statistics for Categorical Features vs. Attrition')
plt.xticks(rotation=90)
plt.show()
#--------------------------------------------------------------------------------------------------------------
category_feature_target_correlation
```

 ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

___Observations :___

_The top 5 columns that have a high relationship with the target ("Attrition") are listed below._

- _"OverTime" (87.56): Working extra hours significantly increases the chance of employees leaving the company._

- _"JobRole" (86.19): Certain job roles have a strong association with employees leaving, possibly because of differences in workload or job satisfaction._

- _"JobLevel" (72.53): The level of a person's job affects the likelihood of them leaving. Higher or lower job levels may influence this._

- _"StockOptionLevel" (60.60): The number of stock options given to employees has some impact on whether they stay or leave, although it's not as strong as other factors._

- _"MaritalStatus" (46.16): Whether someone is married or not slightly affects their likelihood of leaving their job, but it's not as significant as other factors._



_`Meanwhile, the variables "Over 18," "Performance Rating", "Gender," and "Education" exhibit weaker correlations with the target.`_

##### ___2.1.1 Selecting Numerical Features :___
- _Optimal Numerical features can be selected  manually based on a correlation matrix using threshold._

```python
#Function for 
def corr_features(corr_matrix ,threshold):
    correlated_feature_pairs_list= []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if (corr_matrix.iloc[i, j]) >=threshold:
                correlated_feature_pairs_list.append((corr_matrix.columns[i], corr_matrix.columns[j]))
    return correlated_feature_pairs_list
#------------------------------------------------------------------------------------------------
corr_matrix = df[filtered_numerical_columns].corr()
threshold = 0.6
correlated_feature_pairs_list=corr_features(corr_matrix ,threshold)
print(f"These numerical variable pairs exhibit a strong correlation, with a correlation coefficient exceeding {threshold}.")
correlated_feature_pairs_list
```

 ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

___Observations :___

- _Observing the "correlated_feature_pairs_list" reveals the presence of multicollinearity among the features._

- _Columns such as "Age","MonthlyIncome" and "YearsInCurrentRole" are selected from the correlated pairs by comparing with among its pairs and also with the target variable(Point-Biserial Correlation)._

- _The columns "EmployeeCount," "EmployeeNumber," and "StandardHours" have been excluded from consideration for machine learning as they do not contribute significantly to generating meaningful insights._

++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

### ___2.2 Model Selection and Training :___

```python
#Importing dependencies
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,f1_score
```

##### ___2.2.1 Selecting Feature Variables and Target Variable :___

```python
#Feature Variables (X)
X=df_filtered.drop("Attrition",axis=1)
#Target Variable (y)
y=df_filtered["Attrition"]
#------------------------------------------------------------------------------------------------------------------
print("Shape of X :",X.shape)
print("Shape of y :",y.shape)
```

##### ___2.2.2 Train Test Split :___
- _Splitting the dataset for training and testing_

```python
# Split the datasets for training and testing 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y,random_state=7) 
```

##### ___2.2.3 Scaling the Data :___
- _Scaling the data using Standard Scaler_

```python
#Scaling the data using Standard Scaler
sc = StandardScaler()
#--------------------------------------------------
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

##### ___2.2.4 Dimension Reduction :___
- _Trying to reduce Dimension using Principal Component Analysis (PCA)_

```python
#PCA
#Initially ,assign the number of components to "None"
temp_pca= PCA(n_components=None)
X_train_pca =temp_pca.fit_transform(X_train_scaled)
X_test_pca=temp_pca.transform(X_test_scaled)
#-----------------------------------------------------------------------
#Variance ratio 
explained_variance_ratio = temp_pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
#-----------------------------------------------------------------------
#Visualize the Feature size importance
plt.figure(figsize=(22,4))
plt.plot(range(1,44),cumulative_explained_variance,color="blue",marker='o', linestyle='-', markerfacecolor='#00e5ff')
plt.xticks(range(1,44))
plt.title('Principal Componet Analysis (PCA)')
plt.xlabel('Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

 ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

___Observations :___

- _After plotting the cumulative explained variance ratio against the number of components, we can observe that adding more components doesn't significantly increase the cumulative explained variance after around 30 components. This suggests that retaining "30 components" may be sufficient to capture a significant amount of variance in the data without adding too much complexity._

`n_components=30`

++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

```python
#PCA
#Assigning the number of components to "30"
pca= PCA(n_components=30)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca= pca.transform(X_test_scaled)
```

##### ___2.2.5 Training Different Classification Models:___


_Different classification models,such as_

_1. `Logistic Regression`_

_2. `Decision Tree Classifier`_

_3. `Random Forest Classifier`_

_4. `CatBoost Classifier`_

_5. `XGBoost Classifier`_

_6. `Support Vector Classifier`_

_7. `K Nearest Neighbors Class`_

_are trained and checked for performance._

++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

```python
models = {
            "Logistic Regression": LogisticRegression(multi_class='auto'),
            "Decision Tree": DecisionTreeClassifier(random_state=7),
            "Random Forest": RandomForestClassifier(random_state=7),
            "CatBoost": CatBoostClassifier(random_state=7),
            "XGBoost": XGBClassifier(),
            "SVC": SVC(),
            "KNN": KNeighborsClassifier(n_neighbors=5)
         }
#------------------------------------------------------------------------------------------
metrics = []
#------------------------------------------------------------------------------------------
for name, model in models.items():
    model.fit(X_train_pca, y_train)
    y_train_pred = model.predict(X_train_pca)
    y_test_pred = model.predict(X_test_pca)
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred, average='weighted')
    metrics.append({'Model': name, 'Train Accuracy': train_acc, 'Test Accuracy': test_acc, 'F1 Score': f1})
#------------------------------------------------------------------------------------------
metrics_df = pd.DataFrame(metrics)
```

```python
metrics_df
```

```python
# Line Graph for plotting the Model vs Accuracy
plt.figure(figsize=(22,4))
sns.lineplot(data=metrics_df, x="Model", y='Train Accuracy', marker='o', color='#002c8a', label='Train Accuracy')
sns.lineplot(data=metrics_df, x="Model", y='Test Accuracy', marker='o', color='#79a3fc', label='Test Accuracy')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy of Different Models')
plt.legend()
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

 ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

___Observations :___

- _By seeing the performance results,  models like Decision Tree, Random Forest, and XGBoost seem to overfit the training data, others like Logistic Regression, CatBoost, SVC, and KNN offer better generalization performance with relatively balanced Accuracy and F1 Scores._

- _Finally Among all "Logistic Regression" exhibits consistent performance across both train  and test sets, with high Accuracy and a respectable F1 Score._

 _`The F1 Score and Accuracy metrics might show lower performance due to the presence of class imbalance, where certain classes dominate the dataset while others are underrepresented.`_

_`Best performance Model = Logistic Regression`_
  - _Train Acuuracy = 0.89_
  - _Test Accuracy =  0.89_
  - _F1 Score = 0.88_

#### ***2.2.6 Hyperparameter Tuning for Logistic Regression Model :***
- _Using  ***"GridSearchCV"*** for Hyperparameter Tuning_

```python
param_grid = {
            'C': [0.001, 0.01, 0.1, 1, 10, 100],  
            'penalty': ['l1', 'l2'],               
            'solver': ['liblinear', 'saga'],      
            'max_iter': [100, 200, 300],          
             }
#----------------------------------------------------------------------------------------------------
#Model
log_reg_model = LogisticRegression(multi_class='auto')
#----------------------------------------------------------------------------------------------------
#GridSearchCV
grid_search = GridSearchCV(estimator=log_reg_model , param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)
#----------------------------------------------------------------------------------------------------
#Best parameters
best_parameters=grid_search.best_params_
print("1. Best hyperparameters:",best_parameters)
#----------------------------------------------------------------------------------------------------
#Best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_scaled)
#----------------------------------------------------------------------------------------------------
#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("2. Model accuracy:",accuracy)
#----------------------------------------------------------------------------------------------------
#Classification report
report = classification_report(y_test, y_pred)
print("3. Classification Report:")
print(report)
#----------------------------------------------------------------------------------------------------
#Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 5))
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Actual Class")
plt.ylabel("Predicted Class")
plt.show()
```
