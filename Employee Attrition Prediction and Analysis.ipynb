```python
labels = 'Attrition NO','Attrition YES'
colors = ('cyan','crimson')
explode = (0,0.1)
Data['Attrition'].astype(str).value_counts().plot(kind='pie',
                                                 figsize=(15,6),
                                                 autopct='%1.1f%%',
                                                 startangle=90,
                                                 
                                                 labels=None,
                                                 colors=colors,
                                                 explode=explode)

plt.title('Distribution of Employee Attrition in the company', y=1.12)
plt.axis('equal')
plt.legend(labels=labels, loc='upper left')
plt.show()
```
