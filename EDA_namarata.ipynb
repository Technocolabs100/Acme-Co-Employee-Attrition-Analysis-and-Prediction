```python
# Imports:
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data Fetching:

```python
data = pd.read_csv('https://raw.githubusercontent.com/ProPython007/dump/main/preprocessed_preencoded_data.csv')  # pre-processed pre-encoded data (from prev task)
```

##1. Univariate Analysis:

Five-point summary of all the numeric data:

```python
data.describe()
```

Frequency Distribution

```python
fre_loadStatus = data['LoanStatus'].value_counts()
fre_loadStatus
```

##2. Bivariate analysis:

Numerical V/s Categorical

```python
plt.figure(figsize=(15, 10))
sns.barplot(x=data['EmploymentStatus'], y=data['LoanStatus'])
plt.xticks(rotation='vertical')
plt.tight_layout()
```

##3. Multivariate Analysis:

CORRELATION

```python
plt.figure(figsize=(30, 30))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot=True)
```

Pairplot

```python
plt.figure(figsize=(30, 30))
sns.pairplot(data.sample(1000), diag_kind='kde')
```
