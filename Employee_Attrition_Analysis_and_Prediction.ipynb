```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot histograms for numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], bins=20, kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plot box plots for numeric columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, y=column)
    plt.title(f'Box plot of {column}')
    plt.ylabel(column)
    plt.show()
```

```python
# Plot density plots for numeric columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[column], shade=True)
    plt.title(f'Density plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

AS these two columns are useless and don't add any value so, we will remove these two columns

```python
#These fields does not add value, hence removed
df = df.drop(['EmployeeCount','Over18'], axis = 1)
```

```python

# Find duplicate rows based on all columns
duplicate_rows = df[df.duplicated()]

# Print duplicate rows
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicate_rows)
```
